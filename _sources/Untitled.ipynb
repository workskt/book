```python
import rpy2.robjects as ro
from rpy2.robjects.packages import importr, data
from rpy2.robjects import pandas2ri

# pandas変換有効化
pandas2ri.activate()

# Rパッケージ読み込み
ro.r('library(airway)')
ro.r('library(SummarizedExperiment)')  # assay用（重要）

# データロード
ro.r('data("airway")')

# airwayオブジェクト取得
airway_data = ro.r['airway']

# assay関数をR側で直接実行（これが安全）
counts = ro.r('assay(airway)')

# pandas DataFrameに変換
import pandas as pd
counts_df = pandas2ri.rpy2py(counts)

print(counts_df.head())
```

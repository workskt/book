# 文字列

---

## 変数に文字列を代入する


変数に代入した文字列は、一文字取り出したり、逆順にしたりと、いろんな処理を行うことができます。  
DNA配列を文字列として扱う際に便利です。


例として、変数`a`に、`tumor protein p53`という文字列を代入してみます。

````python
a = 'tumor protein p53'
print(a)
````

---

## 特定の文字を取り出す

変数の後ろに`[]`をつけて、その中に数字を入れると、その数字の順番の文字を取り出すことができます。  
この数値は**添字**や**インデックス**と呼ばれます。  
**この順番は0から始まる**ので注意してください。  
最初の文字を取り出したければ、0を入力することになります。

````python
b = a[0]
print(b)
````

インデックスに`0`を指定したので、先頭の`t`が表示されます。

````python
c = a[4]
print(c)
````

インデックスに`4`を指定したので、５文字目の`r`が表示されます。

<br>

**マイナスの数値**を入力すると、後ろから順番に文字を取り出せます。  
−1は最後から1番目の文字、−2は最後から2番目の文字、といった具合です。

````python
d = a[-1]
print(d)
````

インデックスに`-1`を指定したので、最後の文字の`3`が表示されます。

````python
e = a[-3]
print(e)
````

インデックスに`-3`を指定したので、最後から３文字目の`p`が表示されます。

---

## 複数の文字を取り出す

**コロン**`:`を使って、複数の文字を取得することができます。

この操作を**スライス**と呼びます。

<br>

```python
変数[始点:終点]
```

<br>

この時、**終点の1つ前**までを取得することができます。  
ここで入力する順番は、もちろん**最初を0**としたときの順番です。

````python
f = a[6:13]
print(f)
````

7文字目(インデックスでは6)から13文字目(インデックスでは12)までの文字`protein`が表示されます。  
**スペースも一文字とカウント**されます。  
終点で指定した順番の一つ前まで、というのが、慣れるまでは少しややこしいかもしれません。

<br>

始点だけ記入した場合は、その**始点から最後まで**を取得できます。

````python
g = a[6:]
print(g)
````

7文字目以降の`protein p53`が表示されます。

<br>

同様に、終点だけ記載した場合は、**最初から終点の一つ前まで**を取得できます。

````python
h = a[:13]
print(h)
````

13文字目(インデックスでは12)までの`tumor protein`が表示されます。

<br>

`[::-1]`と書くと**逆順**の文字列を取得できます。

````python
r = a[::-1]
print(r)
````

## 文字列を検索する

`re`モジュールの`search()`関数を使うと、指定した文字列を検索することができます。<br>
`re`モジュールはデフォルトでは使用できないため、使う前に`import`で呼び出す必要があります。

````python
import re
````

特に何も出力されませんが、これで`re`モジュールの関数が使用できるようになります。<br><br>
モジュール内の関数を使用するためには、**モジュール名と関数名をカンマで繋ぎます**。<br>
`search()`関数では二つの引数が要求されます。<br>
一つ目に検索したい文字列を、二つ目に検索の対象となる文字列もしくは文字列が格納された変数を指定します。<br><br>
例えば、`tumor protein p53`が格納された変数`a`の中に、`p53`という文字列があるか検索したい場合には以下のように書きます。

````python
match = re.search('p53', a)
print(match)
````

`search()`はマッチする文字が存在した場合、マッチオブジェクトを返します。<br>
ここではマッチオブジェクトを`match`という変数に格納し、それを`print()`で表示させています。<br><br>
返された結果は、index14から17(つまり15文字目から17文字目)に`p53`という文字列がマッチしたことを示しています。<br>
マッチオブジェクト自体はそれほど重要ではないので、あまり気にする必要はありません。

````python
match = re.search('p63', a)
print(match)
````

マッチする文字列が存在しない場合は`None`が返されます。

## 条件と一致する文字列を検索する

特定の文字列だけでなく、条件に合致した文字列を検索することができます。<br>
条件を指定する場合には**正規表現**というものを使います。<br>
正規表現には以下のようなものがあります。<br>

`\d` 数字<br>
`\w` 文字と`_`<br>
`.` 任意の一文字<br>
`*` 直前のパターンを0回以上繰り返し<br>
`+` 直前のパターンを1回以上繰り返し<br>

上記は一例ですので、より詳細は以下のサイトを参照してください。<br>
[Pythonでの正規表現のメタ文字・特殊シーケンスと注意点](https://note.nkmk.me/python-re-match-search-findall-etc/#:~:text=%E6%AD%A3%E8%A6%8F%E8%A1%A8%E7%8F%BE%E3%80%81%E6%96%87%E5%AD%97%E6%95%B0%EF%BC%89-,Python%E3%81%A7%E3%81%AE%E6%AD%A3%E8%A6%8F%E8%A1%A8%E7%8F%BE%E3%81%AE%E3%83%A1%E3%82%BF%E6%96%87%E5%AD%97%E3%83%BB%E7%89%B9%E6%AE%8A%E3%82%B7%E3%83%BC%E3%82%B1%E3%83%B3%E3%82%B9%E3%81%A8%E6%B3%A8%E6%84%8F%E7%82%B9,-Python3%E3%81%AEre)

例えば、`pの後に数字`という文字列が存在するか検索する場合は以下のように書きます。

````python
match = re.search('p\d+', a)
print(match)
````

---
## 条件と一致する部分を抽出する

`group()`メソッドを使うことで、文字列の中から条件に一致する部分だけ取得することができます。<br><br>

以下は、`pの後に数字`という文字列が存在した場合、その該当部分を抽出する方法です。<br>
`match()`を使用する場合は、`search()`で検索する際に**抽出したい部分を( )で囲っておく**必要があります。


````python
match = re.search('(p\d+)', a)
print(match.group(1))
````

条件に該当する`p53`という部分だけ出力されました。<br><br>
`group()`の括弧内の数値は指定した該当部分の1番目を抽出するという命令です。<br>
以下のように、抽出したい部分は複数指定することが可能です。

````python
match = re.search('tumor (protein) (p\d+)', a)
print(match.group(1))
print(match.group(2))
````

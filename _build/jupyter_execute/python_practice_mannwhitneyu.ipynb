# 統計検定を行う
---

データフレームからデータを抽出し、2群間に統計的に有意な差があるか検定を行います。<br>
今回はMann-WhitneyのU検定を行います。<br>
まずはirisデータを取得しましょう。(「pandasデータフレームの扱い方」参照)

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.loc[df['target'] == 0, 'target'] = 'setosa'
df.loc[df['target'] == 1, 'target'] = 'versicolor'
df.loc[df['target'] == 2, 'target'] = 'virginica'
```

```python
df
```

まずはsetosaのsepal length (cm)のデータを取得し、`setosa`という変数に格納します。

```python
setosa = df.loc[df['target'] == 'setosa']['sepal length (cm)']
```

```python
setosa
```

`df.loc[df['target'] == 'setosa']`でsetosaの行だけ取得し、さらにそこに`['sepal length (cm)']`を繋げることでsepal lengthの列を抽出しています。<br>
<br>
virginicaのsepal lengthも同様の方法で取得します。

```python
virginica = df.loc[df['target'] == 'virginica']['sepal length (cm)']
```

両者の中央値に差があるかどうかをMann-WhitneyのU検定により検定します。<br>
`scipy`の`stats`モジュールの`mannwhitneyu`を使用します。<br>
`alternative = 'two-sided'`で両側検定を行います。

```python
from scipy import stats
result = stats.mannwhitneyu(setosa, virginica, alternative = 'two-sided')
```

```python
print(result.statistic)
```

```python
print(result.pvalue)
```

統計的有意に差があることがわかりました。

### 演習2

RNA-seqで20,000遺伝子の発現量変化を解析したところ、2,000遺伝子が発現上昇していました。<br>
GO term XはRNA-seqで解析した20,000遺伝子のうち、100遺伝子が有しています。<br>
今回の解析で発現上昇した2,000遺伝子のGO termを調べたところ、30遺伝子がGO term Xを有していました。<br>
    この結果が生じる確率(30遺伝子**以上**がGO term Xを有している確率)を超幾何分布を用いて算出するコードを作成してください。
